```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
    
plt.title("x-y")
plt.xlabel("x")
plt.ylabel("y")
```

```python
x=np.arange(1,11) 
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':')
plt.plot(x,y2,color='y',linewidth=4)
    
plt.title("x-y")
plt.xlabel("x")
plt.ylabel("y") 
plt.grid('true')
```

```python
x=np.arange(1,11) 
y1=2*x
y2=3*x
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y1,color='r',linestyle=':')
plt.subplot(1,2,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.show()


```
